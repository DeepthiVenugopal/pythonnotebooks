```python
from __future__ import division
import pandas as pd

df = pd.read_csv('datasets/outbreaks.csv')
df.head()
```

```python
import numpy as np
df['Location']=np.where(df['Location'].str.contains('Restaurant'),'Restaurant',df['Location'])
```

```python
df['Location']=np.where(df['Location'].str.contains('Private'),'Home',df['Location'])
```

```python
df['Location']=np.where(df['Location'].str.contains('University'),'School/College/University',df['Location'])
```

```python
df['Location']=np.where(df['Location'].str.contains('Child'),'Child Daycare',df['Location'])
```

```python
df['Location']=np.where(df['Location'].str.contains('Catering'),'Catering Service',df['Location'])
```

```python
df['Location']=np.where(df['Location'].str.contains('Hospital'),'Hospital',(np.where(df['Location'].str.contains('Fair'),'Fair/Festival',df['Location'])))
```

```python
df['Location']=np.where(df['Location'].str.contains('Camp'),'Camp',(np.where(df['Location'].str.contains('Jail'),'Prison/Jail',df['Location'])))
```

```python
df['Location']=np.where(df['Location'].str.contains('Banquet'),'Banquet Facility',(np.where(df['Location'].str.contains('Store'),'Grocery Store',df['Location'])))
```

```python
df['Location']=np.where(df['Location'].str.contains('Unknown'),'Unknown',df['Location'])
```

```python
df['Location'].unique()
```

```python
df.groupby('Location')['Illnesses'].sum().sort_values(ascending = False)
```

```python
df.groupby('State')['Illnesses','Hospitalizations','Fatalities'].sum().sort_values(by = 'Illnesses',ascending = False)
```

```python
df_state = pd.pivot_table(df,index = ['State','Species','Year'],values=['Illnesses','Hospitalizations','Fatalities'],aggfunc='sum').sort_values('Illnesses',ascending = False)
```
