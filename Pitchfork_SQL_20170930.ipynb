```python
import pandas as pd
from sqlalchemy import create_engine

dbfile = r'datasets/database.sqlite'
engine = create_engine(r"sqlite:///{}".format(dbfile))

sql = 'SELECT * from reviews'
df = pd.read_sql(sql,engine)
df.head()
```

```python
df_artistlist = df.groupby('artist')['reviewid'].count()
df_artistlist.sort_values(ascending= False,inplace = True)
df_artistlist.head()
```
