```python
import pandas as pd

names = ['Bob','Jessica','Mary','Jon','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]

gradeslist = zip(names,grades,bsdegrees,msdegrees,phddegrees)

df = pd.DataFrame(data=gradeslist,columns = ['Name','Grades','BS','MS','PhD'])
df
```

```python
df['ssns'] = ['867-53-0909','333-22-4444','123-12-1234','777-93-9311','123-12-1243']
df.head()
```

```python
def right(s,amount):
    return s[-amount]

def standardise_ssn(ssn):
    try:
        ssn=ssn.replace('-','')
        ssn=''.join(ssn.split())
        if len(ssn) <9 and ssn!='Mssing':
            ssn='000000000'+ssn
            ssn=right(ssn,9)
    except:
        pass
        
    return ssn
    df.ssn = df.ssn.apply(standardise_ssn)
    df
```
