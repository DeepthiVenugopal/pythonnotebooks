```python
# survivors per passenger class

df[df['survived']==1].groupby('pclass')['survived'].count()
```

```python
# survivors per passenger class,gender

df[df['survived']==1].groupby(['pclass','sex'])['survived'].count()
```

```python
# survivors per passenger class,gender

df[df['survived']==1].groupby(['pclass','sex'])['survived','age'].count()
```

```python
df[(df['survived'] == 1) & (df['age']<10)].groupby(['pclass','parch'])['survived'].count()
```

```python
#percentage of survivors in each class

100*((df[df['survived']==1].groupby('pclass')['survived'].count())/(df.groupby('pclass')['name'].count()))
```

```python
#percentage of survivors in each class,gender

df_survivor_pct = 100*((df[df['survived']==1].groupby(['pclass','sex'])['survived'].count())/(df.groupby(['pclass','sex'])['name'].count()))
df_survivor_pct
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

df_survivor_pct.plot(kind='bar');plt.title='Titanic Survivors'


```

```python
import statsmodels.formula.api as sm

result  = sm.ols(formula = 'survived ~ pclass -1',data=df).fit()
result.summary()
```
