```python
import pandas as pd
from sqlalchemy import create_engine

dbfile = r'datasets/database.sqlite'
engine = create_engine(r"sqlite:///{}".format(dbfile))

#sql="select name from sqlite_master where type='table';"
reviewgenre_sql = 'SELECT genres.reviewid,genre,title,artist,url,score,best_new_music,author,author_type,pub_date,pub_weekday,pub_day,pub_month,pub_year from genres,reviews where genres.reviewid = reviews.reviewid;'
df_reviewgenre = pd.read_sql(reviewgenre_sql,engine)
genresql = 'select * from genres;'
df_genre = pd.read_sql(genresql,engine)
df_reviewgenre
df_genre
```
